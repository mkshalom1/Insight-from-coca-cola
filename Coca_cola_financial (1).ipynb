```python
# Exploring the total revenue over the years

rev_data = df[df['statement of income'] == 'Revenues'].iloc[0, 1:]

plt.figure(figsize=(12, 6))
plt.plot(rev_data.index, rev_data.values, marker='o', color = 'blue')
plt.title('Total Revenue Between 2011 - 2023')
plt.xlabel('Year')
plt.ylabel('Revenue (in million USD)')
plt.show()
```

```python
# The total revenue of coca cola experienced steady decrease from 2011 till 2020. However, from 2020, they experienced upward growth the revenue till last year (2023).
```

```python
# Explore the gross profit over the years

gross_profit = df[df['statement of income'] == 'GROSS PROFIT'].iloc[0, 1:]

plt.figure(figsize=(12, 6))
plt.plot(gross_profit.index, gross_profit.values, marker='o', color = 'blue' )
plt.title('Gross Profit Over the Years')
plt.xlabel('Year')
plt.ylabel('Gross Profit (in million USD)')
plt.show()
```

```python
# The gross profit was similar to the revenue. The same growth trend is seen over the period of the dataset.
```

```python
# Let us explore and compare the operating income and the net income

operating_income = df[df['statement of income'] == 'OPERATING INCOME'].iloc[0, 1:]
net_income = df[df['statement of income'] == 'NET INCOME ATTRIBUTABLE TO SHAREOWNERS OF THE COCA-COLA COMPANY'].iloc[0, 1:]

plt.figure(figsize=(12, 6))
plt.plot(operating_income.index, operating_income.values, marker='o', linestyle='-', label='Operating Income')
plt.plot(net_income.index, net_income.values, marker='o', linestyle='-', label='Net Income')
plt.title('Operating Income vs Net Income Over the Years')
plt.xlabel('Year')
plt.ylabel('Amount (in million USD)')
plt.legend()
plt.show()
```

```python
# The operating income was consistently more than the net income between the year covered by the dataset
```

```python
# Let's check the trend of the operating expenses between 2011 and 2023

operating_expenses = df[df['statement of income'] == 'Selling, General and Administrative Expense'].iloc[0, 1:]

# Plot
plt.figure(figsize=(12, 6))
plt.plot(operating_expenses.index, operating_expenses.values, marker='o', color = 'blue')
plt.title('Operating Expenses Over the Years')
plt.xlabel('Year')
plt.ylabel('Operating Expenses (in million USD)')
plt.show()
```

```python
# The operating expenses showed similar trend to the revenue and profit. This shows that there is a correlation between the expenses and the revenue and profit.
```

```python
# How was the income tax over the years for coca cola?

income_taxes = df[df['statement of income'] == 'Income taxes'].iloc[0, 1:]

# Plot
plt.figure(figsize=(12, 6))
plt.plot(income_taxes.index, income_taxes.values, marker='o', color = 'blue')
plt.title('Income Taxes Over the Years')
plt.xlabel('Year')
plt.ylabel('Income Taxes (in million USD)')
plt.show()
```

```python
# The income tax trend showed a hike in tax rate in 2017, which seems to be back to normal in the following years . This could as a result
# of the TCJA action, lowering the top statutory corporate tax rate from 35 percent to 21 percent.
```

```python
# Let's explore the trend of interest income versus interest expense

interest_income = df[df['statement of income'] == 'Interest income'].iloc[0, 1:]
interest_expense = df[df['statement of income'] == 'Interest expense'].iloc[0, 1:]

plt.figure(figsize=(12, 6))
plt.plot(interest_income.index, interest_income.values, marker='o', linestyle='-', label='Interest Income')
plt.plot(interest_expense.index, interest_expense.values, marker='o', linestyle='-', label='Interest Expense')
plt.title('Interest Income vs Interest Expense Over the Years')
plt.xlabel('Year')
plt.ylabel('Amount in million (USD)')
plt.legend()
plt.show()
```

```python
# The interest expense was lower than the interest income as at the year 2014, however, since the year 2015, the interest expense has consistently overshot the interest income
```

```python
# Let's check cost of good sold in relation to the revenue

revenue = df[df['statement of income'] == 'Revenues'].iloc[0, 1:]
cost_o_goods = df[df['statement of income'] == 'Cost of Goods and Services Sold'].iloc[0, 1:]
cogs_percentage = (cost_o_goods / revenue) * 100

plt.figure(figsize=(10, 6))
plt.plot(cogs_percentage.index, cogs_percentage.values, marker='o', linestyle='-')
plt.title('Cost of Goods Sold as Percentage of Revenue')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.show()
```

```python
# The highest cost vs revenue percent was witnessed in 2022.
```

```python
# What is the equity loss over the years?

equity_income = df[df['statement of income'] == 'Equity income (loss) - net'].iloc[0, 1:]


plt.figure(figsize=(12, 6))
plt.plot(equity_income.index, equity_income.values, marker='o', color = 'blue')
plt.title('Equity Income (Loss) Over the Years')
plt.xlabel('Year')
plt.ylabel('Equity Income (Loss) in million USD')

plt.show()
```

```python
# The equity loss has steadily increased since 2015.
```

```python
# What is the difference in the income per shares?
basic_net_income_per_share = df[df['statement of income'] == 'BASIC NET INCOME PER SHARE (in dollars per share)'].iloc[0, 1:]
diluted_net_income_per_share = df[df['statement of income'] == 'DILUTED NET INCOME PER SHARE (in dollars per share)'].iloc[0, 1:]

plt.figure(figsize=(12, 6))
plt.plot(basic_net_income_per_share.index, basic_net_income_per_share.values, marker='o', linestyle='-', label='Basic Net Income Per Share')
plt.plot(diluted_net_income_per_share.index, diluted_net_income_per_share.values, marker='o', linestyle='-', label='Diluted Net Income Per Share')
plt.title('Basic and Diluted Net Income Per Share Over the Years')
plt.xlabel('Year')
plt.ylabel('Net Income Per Share (in dollars)')
plt.legend()
plt.show()
```

```python
# The basic and diluted income per shares have experienced synchronic growth over the years.
```
